```python
df = pd.read_csv('PILOT_TEAM_DATA20190711.csv').fillna(0)
df
```

```python
df.columns
```

```python
for i in df.columns:
    print(i)
```

```python
df['Blconsumption'] = df['L30D_BL_CONSUME']/df['L30D_BL_ALOTTED'] 
df['PNSconsumption'] = df['L30D_PNS_ANS']/df['L30D_PNS'] 
df['Coverage'] = df['L30D_TTL_DURATION'] > 30
df.Coverage = df['Coverage'].astype(int)
df['BlconsumptionT'] = df['Blconsumption']  > .40
df['BlconsumptionT'] = df['BlconsumptionT'].astype(int)
df['PNSconsumptionT'] = df['PNSconsumption']  > .75
df['PNSconsumptionT'] = df['PNSconsumptionT'].astype(int)
df['PNSconsumptionT']
```

```python
df = df.groupby(['L4_FSF_ID', 'L3_FSF_ID','L2_FSF_ID','L1_FSF_ID']).sum()
# df[['L1_TELE_ID','L2_TELE_ID','L3_TELE_ID','L4_FSF_ID', 'PNSconsumptionT','BlconsumptionT']]
df.reset_index()
```

```python
df.groupby(['L1_FSF_ID']).sum()
```

```python
dfl1 = df.groupby(['L2_FSF_ID','L1_FSF_ID']).sum()
dfl1.to_csv('L1data.csv')
dfl1 = dfl1.reset_index()
```

```python
dfl3 = df.groupby(['L4_FSF_ID', 'L3_FSF_ID']).sum()
dfl3 = dfl3.reset_index()
dfl3.to_csv('l3data.csv')
```

```python
dfl4 = df.groupby(['L4_FSF_ID']).sum()
dfl4 = dfl4.reset_index()
dfl4.to_csv('l4data.csv')
```

```python
dfl2 = df.groupby(['L3_FSF_ID','L2_FSF_ID']).sum()
dfl2.to_csv('l2data.csv')
dfl2 = dfl2.reset_index()
```

```python
import csv
a = {}
with open('PILOT_TEAM_DATA20190711.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        a[int(row['L3_TELE_ID'].split('.')[0])] = row['L3_TELE_NAME']
a
```
